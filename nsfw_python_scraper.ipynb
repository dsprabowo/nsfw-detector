```python
import pandas as pd
import requests
import os
import time
import numpy as np
```

```python
raw_folder = os.path.join(os.getcwd(),'raw_data')
classes = os.listdir(raw_folder)
```

```python
classes = [c for c in classes if not 'images']
```

```python
raw_img_folder = os.path.join(raw_folder,'images')
print(f"Folder exists : {os.path.exists(raw_img_folder)}")
if os.path.exists(raw_img_folder) == False:
    os.makedirs(raw_img_folder)
```

```python
for c in classes:
    print(f'Scraping {c} images')
    directory = os.path.join(raw_img_folder,c)
    print(f"Folder exists : {os.path.exists(directory)}")
    if os.path.exists(directory) == False:
        print(f"create folder {directory}")
        os.makedirs(directory)
    url_folder = os.path.join(raw_folder,c)
    urls = [f for f in os.listdir(url_folder) if '.txt']
    urls_filename = os.path.join(url_folder,urls[0])
    test = pd.read_csv(urls_filename, sep=' ', header=None,names =['URL'])
    for i,f in enumerate(test.iloc[np.random.randint(0,len(test),3500)]['URL']):
        print(f'Downloading {i}/10000')
        filename = f.split('/')[-1]
        filepath = os.path.join(directory,filename)
        try:
            with open(filepath, 'wb') as g:
                g.write(requests.get(f).content)
            time.sleep(5)
        except Exception as err:
            print(err)
            time.sleep(30)
            pass
    print(f"{c} images downloaded")
```
